# House Prices - Advanced Regression Techniques
## Predict sales prices using PyCaret

### In this Notebook I will use PyCaret as AutoML Library to Predict House Prices, a famous competition on the Kaggle. Line is [Here](https://www.kaggle.com/c/house-prices-advanced-regression-techniques).

### Goal of the Competition
It is your job to predict the sales price for each house. For each Id in the test set, you must predict the value of the SalePrice variable.

### Metric
Submissions are evaluated on Root-Mean-Squared-Error (RMSE) between the logarithm of the predicted value and the logarithm of the observed sales price. (Taking logs means that errors in predicting expensive houses and cheap houses will affect the result equally.)

But for the sake of Simplicity I will use Root-Mean-Squared-Error (RMSE) as my default Parameter.

### File Descriptions:
* **train.csv** - the training set
* **test.csv** - the test set
* **data_description.txt** - full description of each column, originally prepared by Dean De Cock but lightly edited to match the column names used here
* **sample_submission.csv** - a benchmark submission from a linear regression on year and month of sale, lot square footage, and number of bedrooms

## PyCaret `setup` Preprocesing Data

`setup` function is very much important in PyCaret. It preprocess data such as Imputing Numeric & Categorical Features, Target Imbalnce, Removing Outliers(if any), NOrmalize data, Feature Engineering & Selection. We will see many transformation here.

```python
# Before doing any transformation on data let's see `shape` of trainig and testing data.
print(f'Shape of Training Dataset is :', raw_train_data.shape)
print(f'Shape of Testing Dataset is :', raw_test_data.shape)
```

```python
# Let's  `setup` our data

s = setup(
# Basic settings
    data=raw_train_data, target='SalePrice', session_id=42, train_size=0.85,data_split_shuffle=True,n_jobs=-1,
# Imputing Missing Values
    imputation_type='iterative',
    categorical_imputation='mode',
# Ordinal Encoding
    ordinal_features={
        'ExterCond':['Po','Fa','TA','Gd','Ex'],'BsmtQual':['NA','Po','Fa','TA','Gd','Ex'],'BsmtCond':['NA','Po','Fa','TA','Gd','Ex'],
        'HeatingQC':['Po','Fa','TA','Gd','Ex'],'KitchenQual':['Po','Fa','TA','Gd','Ex'],'GarageFinish':['NA','Unf','RFn','Fin'],
        'GarageQual':['NA','Po','Fa','TA','Gd','Ex'],'GarageCond':['NA','Po','Fa','TA','Gd','Ex'],'PavedDrive':['N','P','Y']
    },
# Removing Outliers : It allows us to identify and remove outliers from the dataset before training the model. Outliers are identified through PCA linear dimensionality reduction using the Singular Value Decomposition technique.
    remove_outliers=True,
# Normalize Dataet
    normalize=True, # When set to `True` it uses `Z-Score` for normalization
    transformation=True, # When set to `True` it applies powerful transformer to make data more normal/Gaussian like.
    transform_target=True,
# Feature Selection
    #feature_selection=True,
# Remove Multicollinearity
    #remove_multicollinearity=True,
# Igone Low Varaince Categorical Features
    #ignore_low_variance = True,
    
)
```
